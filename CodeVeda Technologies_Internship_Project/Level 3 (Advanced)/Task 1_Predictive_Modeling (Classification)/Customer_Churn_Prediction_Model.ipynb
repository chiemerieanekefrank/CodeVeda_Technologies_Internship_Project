## Level 3 – Task 1: Predictive Modeling (Classification)

This task focuses on building and evaluating classification models to predict customer churn. The workflow includes data preprocessing, feature scaling, training multiple machine learning models, and evaluating their performance using key metrics such as accuracy, precision, recall, and F1-score.

Additionally, hyperparameter tuning is performed using GridSearchCV to optimize model performance. The goal is to identify the most effective model for predicting customer churn and provide insights that can support customer retention strategies.


```python
# Data handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

# Models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Evaluation
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Hyperparameter tuning
from sklearn.model_selection import GridSearchCV
```

```python
# Load datasets
train_df = pd.read_csv('/content/churn-bigml-80.csv')
test_df = pd.read_csv('/content/churn-bigml-20.csv')

# Preview
train_df.head()
```

```python
train_df.info()
```

## Dataset Overview
The dataset contains customer information used to predict whether a customer will churn. It includes both numerical and categorical features such as state, account length, and service usage patterns. The target variable indicates whether a customer has churned or not.

```python
# Initialize Label Encoder
le = LabelEncoder()

# Encode categorical columns
for col in train_df.columns:
    if train_df[col].dtype == 'object':
        train_df[col] = le.fit_transform(train_df[col])
        test_df[col] = le.transform(test_df[col])
```

```python
# Features
X_train = train_df.drop('Churn', axis=1)
y_train = train_df['Churn']

X_test = test_df.drop('Churn', axis=1)
y_test = test_df['Churn']
```

```python
# Initialize scaler
scaler = StandardScaler()

# Scale features
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Feature Scaling
Feature scaling was applied using StandardScaler to ensure that all features contribute equally to the model performance and to improve the efficiency of the algorithms.

```python
## Logistic Regression
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)
```

```python
## Decision Tree
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
```

```python
## Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
```

```python
## Making Prediction
log_pred = log_model.predict(X_test)
dt_pred = dt_model.predict(X_test)
rf_pred = rf_model.predict(X_test)
```

```python
## Evaluating the Models
def evaluate_model(y_test, y_pred, model_name):
    print(f"--- {model_name} ---")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred))
    print("Recall:", recall_score(y_test, y_pred))
    print("F1 Score:", f1_score(y_test, y_pred))
    print("\n")
```

```python
## Run Evaluation to display the models
evaluate_model(y_test, log_pred, "Logistic Regression")
evaluate_model(y_test, dt_pred, "Decision Tree")
evaluate_model(y_test, rf_pred, "Random Forest")
```

## Model Comparison
Three classification models were evaluated: Logistic Regression, Decision Tree, and Random Forest. Logistic Regression showed lower performance in detecting churn customers, with a very low recall. Decision Tree performed better, providing a good balance between precision and recall. However, Random Forest outperformed both models, achieving the highest accuracy, precision, and F1-score.

## Best Model Selection
The Random Forest model was selected as the best-performing model due to its superior accuracy and balanced performance across precision, recall, and F1-score. It is particularly effective for churn prediction as it successfully identifies a large proportion of churn cases while maintaining high prediction accuracy.

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}

# Initialize GridSearch
grid = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    cv=3,
    scoring='f1',
    n_jobs=-1
)

# Fit grid search
grid.fit(X_train, y_train)
```

## Hyperparameter Tuning
GridSearchCV was used to optimize the Random Forest model by testing different combinations of parameters. The best parameters identified were max_depth = 20, min_samples_split = 5, and n_estimators = 100.

```python
## Printing Best Parameters
print("Best Parameters:", grid.best_params_)
```

```python
## Using The Best Model
best_model = grid.best_estimator_

# Predict again
y_pred_best = best_model.predict(X_test)
```

```python
## Evaluate Tuned Model
evaluate_model(y_test, y_pred_best, "Tuned Random Forest")
```

## Tuned Model Performance
After hyperparameter tuning, the Random Forest model showed improved performance. The recall increased from 0.716 to 0.747, indicating better detection of churn customers. The F1-score also improved, reflecting a better balance between precision and recall. Although precision decreased slightly, the overall model performance improved, making the tuned model more effective for churn prediction.

## Final Conclusion
The predictive modeling task involved training and evaluating multiple classification models, including Logistic Regression, Decision Tree, and Random Forest. Among these, the Random Forest model performed best, achieving the highest accuracy and balanced evaluation metrics.

Hyperparameter tuning further improved the model’s performance, particularly in recall and F1-score, which are critical for churn prediction. The final model demonstrates strong capability in identifying customers likely to churn, making it a valuable tool for decision-making and customer retention strategies.
