## Level 1 – Task 2: Exploratory Data Analysis (EDA)

This task analyzes the Iris dataset to identify patterns, relationships, and statistical properties of the features.
The analysis includes:

Summary statistics

Distribution visualization using histograms

Outlier detection using boxplots

Feature relationships using scatter plots

Correlation analysis using a heatmap

These techniques help understand the structure and relationships within the dataset before applying advanced analytical methods.

```python
# Import libraries for data analysis and visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set(style="whitegrid")
```

```python
# Load the iris dataset
df = pd.read_csv('/content/1) iris.csv')

# Display the first 5 rows
df.head()
```

```python
# Display dataset information
# Shows number of rows, columns, and data types
df.info()
```

## Dataset Overview
The dataset contains measurements of iris flowers including sepal length, sepal width, petal length, and petal width. Each observation is classified into one of three species: Setosa, Versicolor, or Virginica.

```python
# Generate statistical summary for numerical columns
df.describe()
```

## Summary Statistics Insights


*   The dataset contains 150 observations across 4 numerical features.
*   Petal measurements show greater variability than sepal measurements.

*   The range of petal length suggests the presence of multiple species groups.





```python
# Check if there are missing values in any column
df.isnull().sum()
```

```python
# Plot histograms for numerical features
df.hist(figsize=(10,8))

# Add a title to the entire figure
plt.suptitle("Distribution of Iris Dataset Features")

# Display the plots
plt.show()
```

## Histogram Insights


*   Sepal Length values mostly range between 5.0 and 7.0, indicating most flowers have moderate sepal lengths.


*   Sepal Width shows a tighter distribution centered around 3.0, suggesting relatively consistent width measurements across species.
*   Petal Length and Petal Width show clear separation patterns, suggesting they may be strong indicators for distinguishing flower species.


*   Petal measurements show multiple peaks, indicating the presence of different groups within the dataset (likely corresponding to species).

Therefore, this shows that Petal measurements appear more useful than sepal measurements for distinguishing iris species.







```python
# Create boxplots for the dataset features
plt.figure(figsize=(10,6))

sns.boxplot(data=df)

# Add title
plt.title("Boxplot of Iris Dataset Features")

# Display the plot
plt.show()
```

## Boxplots Insights


*   Sepal Length shows moderate variation with no extreme outliers.
*   Sepal Width contains a few outliers (visible points outside the whiskers).


*   Petal Length has a large spread, suggesting strong differences between species.
*   Petal Width also shows significant variation across samples.

Therefore, Petal features show greater variability, making them more informative for classification tasks.





```python
# Scatter plot showing relationship between sepal length and petal length
plt.figure(figsize=(8,6))

sns.scatterplot(data=df,
                x="sepal_length",
                y="petal_length",
                hue="species")

# Add title
plt.title("Sepal Length vs Petal Length by Species")

# Show the plot
plt.show()
```

## Scatter Plots Insights


*   Setosa species forms a clearly separated cluster with small petal lengths.
*   Versicolor species occupies the middle range.


*   Virginica species tends to have larger petal lengths and sepal lengths.

Therefore, Petal length strongly separates the species, especially distinguishing Setosa from the others.





```python
# Compute correlation matrix for numerical columns
corr_matrix = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8,6))

sns.heatmap(corr_matrix,
            annot=True,
            cmap="coolwarm")

# Add title
plt.title("Correlation Matrix of Iris Dataset")

# Display heatmap
plt.show()
```

## Correlation Insights


*   The Setosa species forms a distinct cluster with smaller petal lengths.
*   Sepal length has moderate positive correlation with petal features.


*   Sepal width has weaker relationships with other variables.





```python
# Pairplot visualization to explore relationships between all numerical features
# This helps identify patterns and separation between iris species
sns.pairplot(df, hue="species")
```

## Pairplot Insights
The pairplot shows relationships between all feature pairs simultaneously.

Key Observations:


*   Setosa species is clearly separable from the other species.
*   Versicolor and Virginica display partial overlap but can still be distinguished using petal features.


*   Petal dimensions provide the clearest species separation.





## Exploratory Data Analysis (EDA) Conclusion
The exploratory analysis of the Iris dataset revealed several important patterns. Petal length and petal width exhibit strong correlations and significant variability across species, making them key distinguishing features. Visualizations such as scatter plots and pairplots show that the Setosa species forms a clearly separable cluster, while Versicolor and Virginica display slight overlap but remain distinguishable based on petal measurements. Overall, petal-related features provide stronger predictive power than sepal measurements for identifying iris species.
