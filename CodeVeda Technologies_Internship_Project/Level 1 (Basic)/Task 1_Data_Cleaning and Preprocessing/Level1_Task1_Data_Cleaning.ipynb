## Level 1 – Task 1: Data Cleaning and Preprocessing

This task focuses on preparing a raw dataset for analysis by performing essential preprocessing steps.
(House Prediction Dataset) was used for the Cleaning and Preprocessing.

Steps performed:

Loaded the dataset using pandas

Fixed formatting issues caused by space-separated values

Assigned appropriate column names

Checked for missing values

Checked for duplicate records

Verified dataset structure and data types

After inspection, the dataset was found to contain no missing values or duplicate records, confirming that the dataset is clean and ready for analysis.


```python
import pandas as pd
import numpy as np
```

```python
# Import the pandas library for data manipulation
import pandas as pd

# Load the dataset
# sep=r"\s+" tells pandas that values are separated by one or more spaces
# header=None means the dataset currently has no column names
df = pd.read_csv("4) house Prediction Data Set.csv", sep=r"\s+", header=None)

# Display the first 5 rows of the dataset
df.head()
```

```python
# Define the correct column names for the Boston Housing dataset
column_names = [
    "CRIM",     # Crime rate per capita
    "ZN",       # Residential land zoned
    "INDUS",    # Non-retail business acres
    "CHAS",     # Charles River dummy variable (1 if tract bounds river)
    "NOX",      # Nitric oxide concentration
    "RM",       # Average number of rooms per dwelling
    "AGE",      # Proportion of owner-occupied units built before 1940
    "DIS",      # Distance to employment centres
    "RAD",      # Accessibility to radial highways
    "TAX",      # Property tax rate
    "PTRATIO",  # Pupil-teacher ratio
    "B",        # Proportion of Black population (legacy variable)
    "LSTAT",    # Lower status population percentage
    "MEDV"      # Median house value (TARGET VARIABLE)
]

# Assign the column names to the dataframe
df.columns = column_names

# Display the dataset again to confirm changes
df.head()
```

```python
# Display dataset information
# This shows number of rows, columns, and data types
df.info()
```

```python
# Check if any column contains missing values
df.isnull().sum()
```

The dataset contains no missing values.

Therefore no imputation or deletion is required.

```python
# Check for duplicate rows in the dataset
df.duplicated().sum()
```

```python
# Generate summary statistics for numerical columns
df.describe()
```

```python
# Check dataset dimensions
df.shape
```

```python
# Save the cleaned dataset for later analysis
df.to_csv("cleaned_house_data.csv", index=False)
```
